Find the thirteen adjacent digits in the 1000-digit number that have the greatest product. What is the value of this product?

```python
import pandas as pd
import numpy as np
import csv 
import functools
import operator
```

```python
with open('Euler8_Data.txt', 'r' ) as infile:
    data = infile.readlines()
```

```python
s = data[0]
```

```python
s
```

```python
s = s.replace('', " ")
```

```python
s
```

```python
#s = s.strip()
s = s.split()
```

```python
#s = s.replace(" ", ",")
s
```

```python
#s.split()
s_as_ints = [(int(x))for x in s]
s_as_ints
```

```python
s_as_ints[0:13]
```

```python
def gen_13(n):
    return s_as_ints[n : (n+13)]
```

```python
list_13 = gen_13(127)
```

```python
list_13
```

```python
functools.reduce(lambda x, y: x*y, list_13)
```

```python
values = [functools.reduce(lambda x, y: x*y, gen_13(n)) for n in range(1000)]
```
